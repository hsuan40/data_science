- 利用開發者工具觀察,用什麼方式取得資料及資料回應型態

```python
q = '醫務管理'
url = 'https://www.104.com.tw/jobs/search/list?ro=0&kwop=7&keyword={}&order=2&asc=0&page=1&mode=s'.format(q)
headers = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'}
```

```python
res = requests.get(url, headers = headers) #json 格式資料
content = json.loads(res.text) #json字串轉為dict
```

```python
totalCount = content['data']['totalCount'] # 筆數
totalPage = content['data']['totalPage'] #頁數
```

```python
#第一頁 工作職務訊息 依時間排序
for item in content['data']['list']:
    '''一頁20筆'''
    appearDate = item['appearDate'] # 刊登時間
    applyDesc = item['applyDesc'] #目前應徵人數
    jobName = item['jobName']  #工作職務名稱
    salaryDesc = item['salaryDesc'] #月薪
    
    #其他資料請大家練習整理及打印出來

    print(appearDate,applyDesc,jobName, salaryDesc)
```
