```python
import requests
from bs4 import BeautifulSoup
import re
url = 'https://www.1111.com.tw/job-bank/job-index.asp?si=1&ss=s&ks=醫務管理&page=1'
res = requests.get(url)
soup = BeautifulSoup(res.text,'html.parser')
```

```python
datalist = []
for item in soup.select('#jobResult .halfbox li .displayTable'):
    #左欄資料
    i = item.select('.jbInfo .jbInfoin')[0]   #工作所有訊息節點
    title = i.select('h3 a[title]')[0].text   #職位
    company = i.select('h4 a[title]')[0].text #公司名稱
    cat = i.select('.csort a')[0].text  #行業類別
    #text = i.h4.a['title']
    #cat = re.findall('《行業類別》(\w+)' ,text)[0]  #行業類別
    #cat = re.search('《行業類別》(\w+)' ,text).group(1)  #行業類別
    location = i.select('.needs_m .location')[0].text  #工作地點
    
    needs = (i.select('.needs')[0].text).split('|') #['月薪 28,000~40,000元', '經驗不拘', '大學']
    salary = needs[0]    #薪資
    condition = needs[1] #條件
    degree = needs[2]    #學歷
    
    contents = i.select('.jbInfoTxt p')[0].text  #工作內容
    contents = re.sub('\xa0|\t| ','',contents)
    

    conditionlist = i.select('.jbInfoTxt .conditionBox .conditionList') #應徵詳細條件 (有的沒有資料)
    conditionlist = conditionlist[0].text if len(conditionlist)>0 else None 
    
    #右側資料       
    date = item.select('.jbControl .date')[0].text
                          
    data = {
        'title':title,
        'company':company,
        'cat':cat,
        'location':location,
        'salary':salary,
        'degree':degree,
        'condition':condition,
        'contents':contents,
        'conditionlist':conditionlist,
        'date':date
        
    }
        
    print(data)
    datalist.append(data)
```

```python
import requests
from bs4 import BeautifulSoup
import re

def get_job(q, page, fullpage=False):
    '''
    q為查詢關鍵字，例如'醫務管理'
    page為查詢頁數
    '''
    url = 'https://www.1111.com.tw/job-bank/job-index.asp?si=1&ss=s&ks='+q
    res = requests.get(url)
    soup = BeautifulSoup(res.text,'html.parser')
    pageNum = soup.select('#PageFooter .pageNum')[0].text
    pageNum = pageNum.split('/')[1].replace(' ','') #總頁數,str
            
    #解析網頁函數
    def getdata(soup):        
        datalist = []
        for item in soup.select('#jobResult .halfbox li .displayTable'):
            #左欄資料
            i = item.select('.jbInfo .jbInfoin')[0]   #工作所有訊息節點
            title = i.select('h3 a[title]')[0].text   #職位
            company = i.select('h4 a[title]')[0].text #公司名稱
            cat = i.select('.csort a')[0].text  #行業類別
            #text = i.h4.a['title']
            #cat = re.findall('《行業類別》(\w+)' ,text)[0]  #行業類別
            #cat = re.search('《行業類別》(\w+)' ,text).group(1)  #行業類別
            location = i.select('.needs_m .location')[0].text  #工作地點
            needs = (i.select('.needs')[0].text).split('|') #['月薪 28,000~40,000元', '經驗不拘', '大學']
            salary = needs[0]    #薪資
            condition = needs[1] #條件
            degree = needs[2]    #學歷
            contents = i.select('.jbInfoTxt p')[0].text  #工作內容
            contents = re.sub('\xa0|\t| ','',contents)
            conditionlist = i.select('.jbInfoTxt .conditionBox .conditionList') #應徵詳細條件 (有的沒有資料)
            conditionlist = conditionlist[0].text if len(conditionlist)>0 else None 

            #右側資料       
            date = item.select('.jbControl .date')[0].text

            data = {
                'title':title,
                'company':company,
                'cat':cat,
                'location':location,
                'salary':salary,
                'degree':degree,
                'condition':condition,
                'contents':contents,
                'conditionlist':conditionlist,
                'date':date
            }
            print(data)
            datalist.append(data)
        return datalist
    
    #拿取多頁的資料
    alldata=[]
    if fullpage==True:
        last=int(pageNum)+1
        for page in range(1,last):
            url = 'https://www.1111.com.tw/job-bank/job-index.asp?si=1&ss=s&ks='+q+'&page='+str(page)
            res = requests.get(url)
            soup = BeautifulSoup(res.text,'html.parser')
            datalist = getdata(soup)
            alldata.extend(datalist)
    else:
        last=int(page)+1
        data
        for page in range(1,last):
            url = 'https://www.1111.com.tw/job-bank/job-index.asp?si=1&ss=s&ks='+q+'&page='+str(page)
            res = requests.get(url)
            soup = BeautifulSoup(res.text,'html.parser')
            datalist = getdata(soup)
            alldata.extend(datalist)
    
    return alldata                    
```
