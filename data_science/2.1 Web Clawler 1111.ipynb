```python
import pandas as pd
import jieba
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from wordcloud import get_single_color_func #設定詞雲單一顏色
```

```python
df = pd.read_excel('Data/1111.xls')
df.head()
```

```python
sentences = df.title.values
sentences
```

```python
counter = Counter()
for sentence in sentences:
    counter[sentence] +=1
```

```python
counter
```

```python
dict_wordfreq = dict(counter.items())
```

```python
word_cloud = WordCloud(font_path='/Users/hsuan/work/tools/font/simhei.ttf',
              background_color='white',
              #width=200, height=200, margin=2,
              #color_func=get_single_color_func('#b03535'),
              #max_font_size=80,
              )
word_cloud.fit_words(dict_wordfreq)
plt.imshow(word_cloud)
plt.show()
```

```python
word_cloud = WordCloud(font_path='/Users/hsuan/work/tools/font/simhei.ttf',
              background_color='white',
              width=200, height=200, margin=2,
              color_func=get_single_color_func('#b03535'),
              max_font_size=80,
              )
word_cloud.fit_words(dict_wordfreq)
plt.imshow(word_cloud)
plt.show()
```
