```python
import requests
import json

def clawler104(keyword,page):
    '''
    keyword='醫務管理' 關鍵字
    page= 1 int,第頁數
    
    '''
    params = {
    'ro': 0,
    'kwop': 7,
    'keyword': keyword,
    'order': 11,    
    'asc': 0,
    'page': page,
    'mode': 's'    
}
    url = 'https://www.104.com.tw/jobs/search/list?'
    headers = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'}
    res = requests.get(url, headers=headers, params=params) #json 格式資料
    content = json.loads(res.text) #json字串轉為dict
    
    totalCount = content['data']['totalCount'] # 筆數
    totalPage = content['data']['totalPage'] #頁數
    query = content['data']['query'] # 指定參數類別
    print('查詢'+str(keyword)+'類工作職務')
    print('總共'+str(totalCount)+'筆')
    print('總共'+str(totalPage)+'頁')
    print('目前抓取第'+str(page)+'頁資料')
    return content      
```

```python
import time
#for page in range(1,36)
datalist=[]
for i in range(1,37):
    content = clawler104(keyword='醫務管理',page=i)
    data = content['data']['list'] #list每頁20筆字典資料
    datalist.extend(data)
    time.sleep(1)
```
